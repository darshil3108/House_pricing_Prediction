```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns



data_new=pd.read_csv('houses.csv')
data_new.head(5)
```

```python
column_to_drop=['link','property_id','description','furnishDetails','features']
data_cleaned=data_new.drop(columns=column_to_drop)
data_cleaned.info()
data_cleaned.head(5)
```

```python
import numpy as np

def convert_price(price_str):
    if pd.isna(price_str):
        return np.nan

    # Convert to string (in case the input is numeric)
    price_str = str(price_str).replace(',', '').lower()
    numeric_part = pd.to_numeric(price_str.split()[0], errors='coerce')

    if 'lac' in price_str:
        return numeric_part * 1e5
    elif 'crore' in price_str:
        return numeric_part * 1e7
    else:
        return numeric_part

# Apply the function to the 'price' column
data_cleaned['numeric_price'] = data_cleaned['price'].apply(convert_price)

# Fill missing values with the mean of 'numeric_price'
data_cleaned.fillna({'numeric_price': data_cleaned['numeric_price'].mean()}, inplace=True)

data_cleaned.head(5)
print(data_cleaned['numeric_price'].head(5))
```

```python
data_cleaned.fillna({'society':data_cleaned['society'].mode()},inplace=True)
data_cleaned.fillna({'facing':data_cleaned['facing'].mode()},inplace=True)
```

```python
data_cleaned['bedRoom']=pd.to_numeric(data_cleaned['bedRoom'].str.extract('(\d+)')[0],errors='coerce')
data_cleaned['bathroom']=pd.to_numeric(data_cleaned['bathroom'].str.extract('(\d+)')[0],errors='coerce')
```

```python
data_cleaned.drop_duplicates(inplace=True)
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns


df = data_cleaned


x = df[['bathroom', 'bedRoom']]
y = df['numeric_price']


x = x.dropna()
y = y.dropna()


y = y.loc[x.index]


x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=40)


model = LinearRegression()
model.fit(x_train, y_train)

# Make predictions
y_train_pred = model.predict(x_train)
y_test_pred = model.predict(x_test)

# Charts for x_test (bathroom and bedroom)
plt.figure(figsize=(12, 6))

# Histogram for bathroom in x_test
plt.subplot(1, 2, 1)
sns.histplot(x_test['bathroom'], kde=True, color='blue')
plt.title('Distribution of Bathrooms in Test Set')
plt.xlabel('Bathrooms')
plt.ylabel('Frequency')

# Histogram for bedroom in x_test
plt.subplot(1, 2, 2)
sns.histplot(x_test['bedRoom'], kde=True, color='green')
plt.title('Distribution of Bedrooms in Test Set')
plt.xlabel('Bedrooms')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()

# Comparison of x_train, y_train, and predictions
plt.figure(figsize=(12, 6))

# Scatter plot for actual vs predicted values (training set)
plt.subplot(1, 2, 1)
plt.scatter(y_train, y_train_pred, color='blue', alpha=0.5)
plt.plot([y_train.min(), y_train.max()], [y_train.min(), y_train.max()], 'k--', lw=2)  # Diagonal line
plt.title('Training Set: Actual vs Predicted')
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')

# Scatter plot for actual vs predicted values (test set)
plt.subplot(1, 2, 2)
plt.scatter(y_test, y_test_pred, color='red', alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2)  # Diagonal line
plt.title('Test Set: Actual vs Predicted')
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')

plt.tight_layout()
plt.show()
```

```python
print("training set")
plt.scatter(x_train['bedRoom'],y_train,color='red')
plt.plot(x_train['bedRoom'],y_train_pred,color='blue')
plt.xlabel(" bedroom size")
plt.ylabel("Price ")
plt.show()

plt.scatter(x_test['bedRoom'],y_test,color='blue',label="Actual data")
plt.plot(x_test['bedRoom'],y_test_pred,color='red',label="Predicted data")
plt.legend()
plt.xlabel("X test")
plt.ylabel("Y test")
plt.title("Testing set")
plt.show()
```

```python
print("training set")
plt.scatter(x_train['bathroom'],y_train,color='red')
plt.plot(x_train['bathroom'],y_train_pred,color='blue')
plt.xlabel(" bedroom size")
plt.ylabel("Price ")
plt.show()

plt.scatter(x_test['bathroom'],y_test,color='blue',label="Actual data")
plt.plot(x_test['bathroom'],y_test_pred,color='red',label="Predicted data")
plt.legend()
plt.xlabel("X test")
plt.ylabel("Y test")
plt.title("Testing set")
plt.show()
```

```python
y_train_pred = model.predict(x_train)  # Predictions for training set
y_test_pred = model.predict(x_test)  # Predictions for test set

# Sort the data for smooth trend lines
train_sorted_bathroom = x_train.copy()
train_sorted_bathroom['y_train'] = y_train
train_sorted_bathroom['y_train_pred'] = y_train_pred
train_sorted_bathroom = train_sorted_bathroom.sort_values(by='bathroom')  # Sort by 'bathroom'

train_sorted_bedRoom = x_train.copy()
train_sorted_bedRoom['y_train'] = y_train
train_sorted_bedRoom['y_train_pred'] = y_train_pred
train_sorted_bedRoom = train_sorted_bedRoom.sort_values(by='bedRoom')  # Sort by 'bedRoom'

test_sorted_bathroom = x_test.copy()
test_sorted_bathroom['y_test'] = y_test
test_sorted_bathroom['y_test_pred'] = y_test_pred
test_sorted_bathroom = test_sorted_bathroom.sort_values(by='bathroom')  # Sort by 'bathroom'

test_sorted_bedRoom = x_test.copy()
test_sorted_bedRoom['y_test'] = y_test
test_sorted_bedRoom['y_test_pred'] = y_test_pred
test_sorted_bedRoom = test_sorted_bedRoom.sort_values(by='bedRoom')  # Sort by 'bedRoom'

# Plotting for Training Set (Bathroom)
plt.figure(figsize=(10, 6))
plt.scatter(x_train['bathroom'], y_train, color='red', alpha=0.5, s=20, label='Actual Data')
plt.plot(train_sorted_bathroom['bathroom'], train_sorted_bathroom['y_train_pred'], color='blue', linewidth=2, label='Predicted Trend')
plt.xlabel("Bathroom")
plt.ylabel("Price")
plt.title("Training Set: Bathroom vs Price ")
plt.legend()
plt.grid(True)
plt.show()

# Plotting for Training Set (BedRoom)
plt.figure(figsize=(10, 6))
plt.scatter(x_train['bedRoom'], y_train, color='green', alpha=0.5, s=20, label='Actual Data')
plt.plot(train_sorted_bedRoom['bedRoom'], train_sorted_bedRoom['y_train_pred'], color='purple', linewidth=2, label='Predicted Trend')
plt.xlabel("BedRoom")
plt.ylabel("Price")
plt.title("Training Set: BedRoom vs Price ")
plt.legend()
plt.grid(True)
plt.show()

# Plotting for Test Set (Bathroom)
plt.figure(figsize=(10, 6))
plt.scatter(x_test['bathroom'], y_test, color='blue', alpha=0.5, s=20, label='Actual Data')
plt.plot(test_sorted_bathroom['bathroom'], test_sorted_bathroom['y_test_pred'], color='red', linewidth=2, label='Predicted Trend')
plt.xlabel("Bathroom")
plt.ylabel("Price")
plt.title("Testing Set: Bathroom vs Price ")
plt.legend()
plt.grid(True)
plt.show()

# Plotting for Test Set (BedRoom)
plt.figure(figsize=(10, 6))
plt.scatter(x_test['bedRoom'], y_test, color='orange', alpha=0.5, s=20, label='Actual Data')
plt.plot(test_sorted_bedRoom['bedRoom'], test_sorted_bedRoom['y_test_pred'], color='brown', linewidth=2, label='Predicted Trend')
plt.xlabel("BedRoom")
plt.ylabel("Price")
plt.title("Testing Set: BedRoom vs Price ")
plt.legend()
plt.grid(True)
plt.show()
```
